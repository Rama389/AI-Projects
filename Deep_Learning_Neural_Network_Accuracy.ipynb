```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical

# --------------------
# 1. Load dataset
# --------------------
df = pd.read_csv("tmnst Dataset.csv")

# Remove all-zero columns
X = df.drop(columns=["labels"])
X = X.loc[:, (X != 0).any(axis=0)].astype("float32")

# Normalize to [0,1]
X /= 255.0

# Labels
y = df["labels"].values
encoder = LabelEncoder()
y = encoder.fit_transform(y)
y = to_categorical(y, num_classes=26)

# Split into train/test (70/30 stratified)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, stratify=y, random_state=42
)

input_dim = X.shape[1]
output_dim = 26

# --------------------
# 2. Train architectures
# --------------------
architectures = {
    "1-5": [5],
    "1-10": [10],
    "1-20": [20],
    "1-25": [25],
    "2-5-5": [5, 5],
    "2-5-10": [5, 10],
    "2-10-5": [10, 5],
    "2-10-10": [10, 10]
}

results = []

for name, layers in architectures.items():
    model = Sequential()
    # Input + first hidden layer
    model.add(Dense(layers[0], activation="relu", input_shape=(input_dim,)))
    # Other hidden layers
    for units in layers[1:]:
        model.add(Dense(units, activation="relu"))
    # Output layer
    model.add(Dense(output_dim, activation="softmax"))

    model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])

    # Train model
    model.fit(X_train, y_train, epochs=10, batch_size=64, verbose=0)

    # Evaluate
    loss, acc = model.evaluate(X_test, y_test, verbose=0)
    num_params = model.count_params()

    results.append([name, num_params, acc])

# --------------------
# 3. Save results to CSV
# --------------------
results_df = pd.DataFrame(results, columns=["Architecture", "Parameters", "Test_Accuracy"])
results_df.to_csv("nn_results.csv", index=False)

print("✅ Results saved to nn_results.csv")
print(results_df)
```
