```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import statsmodels.api as sm

```

```python
#url = "http://www.stat.columbia.edu/~gelman/arm/examples/police/frisk_with_noise.dat"
df = pd.read_csv("Diesel Engines - Yajvin-Final.csv")
df.head()
```

```python
y = df.Breakdowns
```

```python
#Assigning a column of 1's for intercept, since we are using statmodels.api
#this is not necessary if you are using statmodels.formula.api
df = (df

   .assign(intercept=1)  # Adds a column called 'intercept' with all values equal to 1.

    .reset_index(drop=True)
)
```

```python
df.head()
```

```python
y.head()
```

```python
df= df.drop(columns=["Breakdowns"])
```

```python
df.head()
```

```python
y.head()
```

```python
model_no_indicators = sm.GLM(
    y,
    df["intercept"],

    family=sm.families.Poisson(),
)
result_no_indicators = model_no_indicators.fit()
print(result_no_indicators.summary())
```

```python
#Now use all variables
```

```python
df.head()
```

```python
model_with_all_variables = sm.GLM(
    y,
    df,
    family=sm.families.Poisson(),
)

result_with_all_variables = model_with_all_variables.fit()
print(result_with_all_variables.summary())
```
